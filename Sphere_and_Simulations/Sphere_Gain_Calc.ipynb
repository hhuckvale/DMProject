Use this code to calculate the gain for a certain voltage. 

Use the plot 'amplitude-voltage_plot.png' to find the amplitude corresponding to the voltage at a certain pressure.

```python
voltage = '4000V'
pressure1 = 1 #bar
amplitude_adu = 3000

#convert from adu to V/pc 

V = amplitude_adu * (2.5 / 2**16)

# preamp sensitivity
S = 1.4 #V/pC

# collected charge
Q_pC = V / S
Q_C = Q_pC * 1e-12

#V/pC = GNe, find G
#1 muon ~30keV deposited, N = 30keV/30eV = 1000 electrons per muon produced

energy_dep = 5.5e6 #in eV
N = energy_dep / 30
e = 1.6e-19

gain1 = Q_C / (N*e)

pressure2 = 0.1 #bar

print('pressure:', pressure1)
print('sphere voltage:', voltage)
print('gain:',gain1)
```
