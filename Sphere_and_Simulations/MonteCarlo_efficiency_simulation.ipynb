```python
import numpy as np

N=1000000
square_length = 300 #in cm

x0 = np.zeros(N)
y0 = np.zeros(N)
z0 = np.zeros(N)
theta = np.zeros(N)
phi = np.zeros(N)
dx = np.zeros(N)
dy = np.zeros(N)
dz = np.zeros(N)


#produce muon trajectories

for muon in range(N):
    x0[muon] = np.random.uniform(-square_length/2, square_length/2)
    y0[muon] = np.random.uniform(-square_length/2, square_length/2)
    z0[muon] = 0


    #angles
    accepted = False
    
    while not accepted:
        theta_trial = np.random.uniform(-np.pi/2, np.pi/2)
        n = np.random.uniform(0, 1)

        if n < (2/np.pi) * np.cos(theta_trial)**2:
            theta[muon] = theta_trial
            accepted = True

    phi[muon] = np.random.uniform(0,2*np.pi)
    #theta[muon] = 0
    #directions
    dx[muon] = np.sin(theta[muon])*np.cos(phi[muon])
    dy[muon] = np.sin(theta[muon])*np.sin(phi[muon])
    dz[muon] = -np.cos(theta[muon])
```

```python
#scintillator panels separate by 2cm, each with thickness 1cm
scintillator1_pos = np.array([0,0,0])
scintillator2_pos = np.array([0,0,-3])
scintillator3_pos = np.array([0,0,-6])
scint_width = 25
scint_length = 135
scint_thickness = 1

all_trigger = np.zeros(N, dtype=bool)
top_two_trigger = np.zeros(N,dtype=bool)
bottom_two_trigger = np.zeros(N,dtype=bool)
top_and_bottom_only = np.zeros(N,dtype=bool)

for muon in range(N):
    # intersection with scintillator 1
    t1 = (scintillator1_pos[2] - z0[muon]) / dz[muon]
    x1 = x0[muon] + t1*dx[muon]
    y1 = y0[muon] + t1*dy[muon]

    hit1 = False
    if abs(x1) < scint_width/2 and abs(y1) < scint_length/2:
        hit1 = True

    # intersection with scintillator 2
    t2 = (scintillator2_pos[2] - z0[muon]) / dz[muon]
    x2 = x0[muon] + t2*dx[muon]
    y2 = y0[muon] + t2*dy[muon]

    hit2 = False
    if abs(x2) < scint_width/2 and abs(y2) < scint_length/2:
        hit2 = True

    # intersection with scintillator 3
    t3 = (scintillator3_pos[2] - z0[muon]) / dz[muon]
    x3 = x0[muon] + t3*dx[muon]
    y3 = y0[muon] + t3*dy[muon]

    hit3 = False
    if abs(x3) < scint_width/2 and abs(y3) < scint_length/2:
        hit3 = True

    # coincidence
    if hit1 and hit2 and hit3:
        all_trigger[muon] = True

    if hit1 and hit2 and not hit3:
        top_two_trigger[muon] = True
        
    if hit2 and hit3 and not hit1:
        bottom_two_trigger[muon] = True

    if hit1 and hit3 and not hit2:
        top_and_bottom_only[muon] = True


print('all_trigger:', np.sum(all_trigger))
print('top_two_trigger:', np.sum(top_two_trigger))
print('bottom_two_trigger:', np.sum(bottom_two_trigger))
print('top and bottom only, not middle:', np.sum(top_and_bottom_only))
```
