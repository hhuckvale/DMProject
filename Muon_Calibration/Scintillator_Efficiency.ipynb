This code evalutates the efficiency of the scintillator detector. 

```python
import matplotlib.pyplot as plt # https://matplotlib.org/
import numpy as np # http://www.numpy.org/
import tekwfm
import os
from scipy import integrate
from scipy.optimize import curve_fit
import pandas as pd
from scipy.stats import moyal as moyal_dist
```

```python
#import files - 
doVerbose = False

#import the wfm file
df_ch1_clean_OR = pd.read_csv('-400mV_trigger_clean/240226_muoncal_20000_1900_4_on_ch1_proc_clean.csv', index_col=0)
df_ch1_clean_AND = pd.read_csv('-400mV_trigger_clean/170226_muoncal_20000_1900_1_on_ch1_proc_clean.csv', index_col=0)

df_ch2_clean_OR = pd.read_csv('-400mV_trigger_clean/240226_muoncal_20000_1900_4_on_ch2_proc_clean.csv', index_col=0)
df_ch2_clean_AND = pd.read_csv('-400mV_trigger_clean/170226_muoncal_20000_1900_1_on_ch2_proc_clean.csv', index_col=0)

datasets= {
    "ch1_OR": df_ch1_clean_OR,
    "ch1_AND": df_ch1_clean_AND,
    "ch2_OR": df_ch2_clean_OR,
    "ch2_AND": df_ch2_clean_AND
}
```

```python
# find duration of data collection
durations = {}
for name, df in datasets.items():
    t_start = df["event_timestamps"].min()
    t_end = df["event_timestamps"].max()
    durations[name] = t_end - t_start

# find minimum duration
# DCT = Data collection time
DTC_ch1_OR = durations["ch1_OR"]
DTC_ch1_AND = durations["ch1_AND"]
DTC_ch2_OR = durations["ch2_OR"]
DTC_ch2_AND = durations["ch2_AND"]

min_DTC = min(durations.values())
print(min_DTC)
```

```python
# Restrict each dataset to the min_DTC window
restricted_dfs = {}
for name, df in datasets.items():
    t_start = df["event_timestamps"].min()
    cutoff = t_start + min_DTC
    times_in_window = df["event_timestamps"] <= cutoff
    restricted_dfs[name] = df[times_in_window].copy()

# call restricted datasets
clean_ch1_OR = restricted_dfs["ch1_OR"]
clean_ch1_AND = restricted_dfs["ch1_AND"]
clean_ch2_OR = restricted_dfs["ch2_OR"]
clean_ch2_AND = restricted_dfs["ch2_AND"]

# obtain number of events in each dataset
count_ch1_OR = len(clean_ch1_OR)
count_ch1_AND = len(clean_ch1_AND)
count_ch2_OR = len(clean_ch2_OR)
count_ch2_AND = len(clean_ch2_AND)

Rate_ch1_OR = len(df_ch1_clean_OR) / DTC_ch1_OR
Rate_ch1_AND = len(df_ch1_clean_AND) / DTC_ch1_AND
Rate_ch2_OR = len(df_ch2_clean_OR) / DTC_ch2_OR
Rate_ch2_AND = len(df_ch2_clean_AND) / DTC_ch2_AND

print(count_ch1_AND)
print(count_ch2_OR)
```

```python
# Efficiency of Top scintillator
#eff_2 = count_ch2_AND / count_ch1_OR
eff_2 = Rate_ch2_AND / Rate_ch1_OR
print(eff_2)

# Efficiency of Bottom scintillator
#eff_1 = count_ch1_AND / count_ch2_OR
eff_1 = Rate_ch1_AND / Rate_ch2_OR
print(eff_1)
```
